```python
# Annotating with text
# plt.style.use('fivethirtyeight') #hoping this changes the font

fig, ax = plt.subplots()

# fine-tuning of title
fake.plot.bar(width=0.9, color='#d46108') # custom color to match
ax.set_title("'An Inconvenient Sequel: Truth to Power' is divisive",
             fontsize=13,
             fontweight='bold',
             x=0.41,
             y=1.15)
# subtitle
ax.text(s='IMDB Ratings for the film as of Aug. 29th',
        fontsize=11,
        fontweight='light',
        x=-1.6,
        y=44)

# Labeling of axes
ax.set_ylabel("Percent of total votes", fontsize=9, fontweight='bold')
ax.set_xlabel("Rating", fontsize=9, fontweight='bold')

# ytick labeling
yticks = [0, 10, 20, 30, 40]
ytickslables = ['0  ', '10  ', '20  ', '30  ', '40%'] # whitespace for alignment

ax.set_yticks(yticks)
ax.set_yticklabels(ytickslables,
                   fontsize=10,
                   fontweight='ultralight',
                   verticalalignment='center',
                   color='grey')
# xtick labeling
ax.set_xticks(range(0,11)) # This took longer than it should have to recognize
ax.set_xticklabels(range(1,11), fontsize=10,
                   fontweight='ultralight',
                   rotation='horizontal',
                   color='grey')

# background color matching
fig.patch.set(color='white')
ax.patch.set(color='white')

# this gives those nice extended gridlines
ax.set_xlim(left=-1)
ax.set_ylim(top=40.5, bottom=-2)

# spine settings
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.grid(linewidth=1)
fig.show()
```

## Recreating a second graph from 5ThirtyEight

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

cable_news = [3.1,3.7,2.8,3.2,2.5,3.7,4.1,3.0,
              3.6,2.3,1.2,2.2,1.6,1.9,2.0,1.9,
              1.6,1.9,2.2,1.7,2.6,4.7,5.0,1.6,1.7]

online_news = [1.1,1.2,1.3,1.3,1.05,1.1,1.5,1.3,1.4,
               1.7,1.1,1.05,1.25,0.9,0.9,1.05,
               1.2,0.9,0.85,1.1,1.1,4.7,3.1,1.4,1.5]
df = pd.DataFrame(data={'cable_news':cable_news, 'online_news':online_news})

fig, ax = plt.subplots()
df['cable_news'].plot(linewidth=3, color='#03e3fc')
df['online_news'].plot(linewidth=3.5, color='#d46108')
# fine-tuning of title
 # custom color to match
ax.set_title("Biden's Ukraine-related media bump is fading",
             fontsize=13,
             fontweight='bold',
             x=0.41,
             y=1.15)
# subtitle
ax.text(s="Cable news",
        fontsize=9,
        fontweight='bold',
        x=0.3,
        y=3.8)

# assorted text
ax.text(s="Online news",
        fontsize=9,
        fontweight='bold',
        x=0.3,
        y=1.6)

ax.text(s="After the Ukraine news broke,\nBiden was mentioned " \
 + r"$\bf{" + 'five' + "}$" + ' ' + r"$\it\bf{" + 'times' + "}$" + \
 "\nmore than the next most-mentioned'\n'candidate on cable news",
        fontsize=9,
        fontweight='normal',
        fontstyle='italic',
        x=20,
        y=4.0,
        ha='right')

ax.text(s="Biden's share of media mentions on each medium relative \
to the next\nmost-mentioned candidate each week",
        fontsize=11,
        fontweight='normal',
        x=-4,
        y=5.5)


# plot has no axis labels

# ytick labeling

ytickslables = ["EVEN", '2', '3', '4', '5X'] # whitespace for alignment

ax.set_yticks(range(1,6))
ax.set_yticklabels(ytickslables,
                   fontsize=10,
                   fontweight='ultralight',
                   verticalalignment='center',
                   horizontalalignment='right',
                   color='grey')
# xtick labeling
xticks = [0, 4.5, 9, 13.5, 18, 22.5, 27]
ax.set_xticks(xticks) # This took longer than it should have to recognize
xticklabels = ["MAY '19", "JUNE", "JULY", "AUG.", "SEPT.", "OCT."]
ax.set_xticklabels(xticklabels, fontsize=10,
                   fontweight='ultralight',
                   rotation='horizontal',
                   color='grey')
# stolen from a demo of arrows
ax.annotate('',
            xy=(22, 5.05), xycoords='data',
            xytext=(-60, 0), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=40,angleB=-90"))


# background color matching
fig.patch.set(color='#e8e8e8')
ax.patch.set(color='#e8e8e8')

# this gives those nice extended gridlines
ax.set_xlim(left=-1.5, right=25)
ax.set_ylim(top=5.2, bottom=0.7)

# spine settings
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.grid(linewidth=1)
gridlines = ax.yaxis.get_gridlines()
gridlines[0].set_linewidth(2.5)
gridlines[0].set_color('grey')
#print(dir(gridlines[0]))
fig.show()
```
